```python
import geopandas as gpd
import pandas as pd
```

```python
aichi_ku = gpd.read_file('N03-23_23_230101.geojson')
```

```python
aichi_ku.plot()
```

```python
aichi_ku.info()
```

```python
aichi_ku_pop = pd.read_csv('aichi_ku.csv')
```

```python
aichi_ku_pop.info()
```

```python
aichi_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
aichi_ku
```

```python
aichi_ku['KEY_CODE'] = aichi_ku['KEY_CODE'].astype(int)
```

```python
aichi_ku.info()
```

```python
aichi_ku_pop = aichi_ku.merge(aichi_ku_pop,on='KEY_CODE')
```

```python
aichi_ku_pop
```

```python
aichi_ku_pop.info()
```

```python
aichi_ku_pop['総数'] = aichi_ku_pop['総数'].astype(int)
```

```python
aichi_ku_pop.info()
```

```python
aichi_ku_pop.plot(column='総数',legend=True)
```
